```python
try:
    a = int(input("Enter the number: "))
    b = 7 / a
    print("Ans: ", b)
except ValueError as e: 
    print("error is occurs: -- ", e)
finally : 
    print("This will be executed always!!")

print("Code end here")
```

```python
try:
    a = int(input("Enter the number: "))
    b = 7 / a
    print("Ans: ", b)
except ValueError as e: 
    print("error is occurs: -- ", e)
except ZeroDivisionError as e: 
    print("error is occurs: -- ", e)
except Exception as e: 
    print("error is occurs: -- ", e)
finally : 
    print("This will be executed always!!")

arr = [1,3,6,2,4,9,10]
try: 
    for num in arr:
        print(num)
except Exception as e:
    print("error is occurs:-- ", e)
print("Code end here")
```
